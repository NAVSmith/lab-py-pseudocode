### Exercise 2: Check Even or Odd

Write pseudocode that checks if a number is even or odd and prints the result. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Add your code solution here
# Pseudocode:
# 1. Start
# 2. Read a number and store it in variable num
# 3. If num modulo 2 equals 0, then
#    a. Print "num is even"
# 4. Else
#    a. Print "num is odd"
# 5. End
# Python Code:
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(num, "is even")
else:
    print(num, "is odd")

```

### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Add your code solution here

# Pseudocode:
# 1. Start
# 2. Read a number N and store it in variable N
# 3. Initialize sum to 0
# 4. For i from 1 to N, do the following:
#    a. Add i to sum
# 5. Print the value of sum
# 6. End
# Python Code:
N = int(input("Enter a number N: "))
sum = 0
for i in range(1, N + 1):
    sum += i
print("The sum of the first", N, "natural numbers is:", sum) 
```

### Exercise 4: Find Maximum of Three Numbers

Write pseudocode that finds the maximum of three numbers entered by the user. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Add your code solution here
#Write pseudocode that finds the maximum of three numbers entered by the user. Then, write the equivalent Python code.
# Pseudocode:
# 1. Start
# 2. Read three numbers and store them in variables num1, num2, and
#    num3
# 3. If num1 is greater than or equal to num2 and num3, then
#    a. Set max_num to num1
# 4. Else if num2 is greater than or equal to num1 and num3, then
#    a. Set max_num to num2
# 5. Else
#    a. Set max_num to num3
# 6. Print the value of max_num
# 7. End
# Python Code:
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
if num1 >= num2 and num1 >= num3:
    max_num = num1
elif num2 >= num1 and num2 >= num3:
    max_num = num2
else:
    max_num = num3
print("The maximum of the three numbers is:", max_num)
```

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python

# Pseudocode:
# 1. Start
# 2. Read a string and store it in variable text
# 3. Initialize vowel_count to 0
# 4. For each character char in text, do the following:
#    a. If char is a vowel (a, e, i, o, u, A, E, I, O, U), then
#       i. Increment vowel_count by 1
# 5. Print the value of vowel_count
# 6. End
# Python Code:
text = input("Enter a string: ")
vowel_count = 0
for char in text:
    if char.lower() in 'aeiou':
        vowel_count += 1
print("The number of vowels in the string is:", vowel_count)
```

### Bonus Exercise 6: Reverse a String

Write pseudocode to reverse a given string and print it out.

<!-- Add your pseudocode here -->

```python
# Add your code solution here

# Pseudocode:
# 1. Start
# 2. Read a string and store it in variable text
# 3. Initialize reversed_text to an empty string
# 4. For each character char in text (from the last character to the first), do the following:
#    a. Append char to reversed_text
# 5. Print the value of reversed_text
# 6. End
# Python Code:
text = input("Enter a string: ")
reversed_text = ""
for char in text[::-1]:
    reversed_text += char
print("The reversed string is:", reversed_text)
```

### Bonus Exercise 7: Print Multiplication Table of a Number

Write pseudocode to print the multiplication table of a given number up to 10.

<!-- Add your pseudocode here -->

```python
# Add your code solution here

# Pseudocode:
# 1. Start
# 2. Read a number and store it in variable num
# 3. For i from 1 to 10, do the following:
#    a. Calculate product as num multiplied by i
#    b. Print "num x i = product"
# 4. End
# Python Code:
num = int(input("Enter a number: "))
for i in range(1, 11):
    product = num * i
    print(f"{num} x {i} = {product}") 
```
